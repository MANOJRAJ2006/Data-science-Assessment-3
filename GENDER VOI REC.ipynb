# Basic libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import homogeneity_score, confusion_matrix
```

```python

import warnings
warnings.filterwarnings("ignore")
```

# Load the dataset

```python
data = pd.read_csv("voice.csv")
```

```python
data
```

```python
X_features = data.drop('label', axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_features)
```

```python
X_scaled.shape
```

```python
inertia = []
k_range = range(1,10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=100)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=100)
kmeans.fit(X_scaled)
```

```python
cluster_labels = kmeans.labels_
```

```python
true_labels = data['label'].map({'male': 0, 'female': 1})
```

```python
cluster_mapping_check = pd.DataFrame({'True_Label': true_labels, 'Cluster': cluster_labels})
mapping = cluster_mapping_check.groupby('Cluster')['True_Label'].mean()
```

```python
if mapping.iloc[0] < mapping.iloc[1]:
     predicted_labels = cluster_labels
else:
     predicted_labels = 1 - cluster_labels
```

```python
homogeneity = homogeneity_score(true_labels, cluster_labels)
```

```python
cm = confusion_matrix(true_labels, predicted_labels)
accuracy = np.sum(np.diag(cm)) / np.sum(cm)
```

```python
cluster_centers_scaled = kmeans.cluster_centers_
```

```python

cluster_centers_original = scaler.inverse_transform(cluster_centers_scaled)
centers_df = pd.DataFrame(cluster_centers_original, columns=X_features.columns)
```

```python
if mapping.iloc[0] < mapping.iloc[1]:
    centers_df.index = ['Male Cluster Centroid', 'Female Cluster Centroid']
else:
    centers_df.index = ['Female Cluster Centroid', 'Male Cluster Centroid']
```

```python
key_features_df = centers_df[['meanfun', 'IQR', 'sd']].transpose()
key_features_df['Difference'] = key_features_df['Female Cluster Centroid'] - key_features_df['Male Cluster Centroid']
print(key_features_df)
```
